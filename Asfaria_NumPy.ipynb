# Matplotlib
Matplotlib acts like a state machine. Any command is applied to current plotting area.

````python
import matplotlib.pyplot as plt
import numpy as np
t = np.linspace(0,2*np.pi,50)
x = np.sin(t)
y = np.cos(t)

# Create a fig
plt.figure()
# Plot x inside it
plt.plot(x)

# Create a NEW fig
plt.figure()
# Plot y inside it
plt.plot(y)
# Title it
plt.title("cos")
````

# NumPy: Array Broadcasting

NumPy arrays of different dimensionality can be combined in the same expression.
******
Arrays with smaller dimension are **broadcasted** to match the larger arrays, *without copying data*.
******
Broadcasting has two rules:
1. Prepend ones to smaller array's shape
2. Dimensions of size 1 are repeated without copying

## Rule 1: Prepend Ones to Smaller Array's Shape

````python
import numpy as np
a = np.ones((3,5))
# a.shape == (3,5)
````

````python
print(a)
````

````python
b = np.ones((5,))
# b.shape == (5,)
````

````python
print(b)
````

````python
b.reshape(1,5)
# result is a (1,5) -shaped array
````

````python
b[np.newaxis, :]
# equivalent, more concise
````

## Rule 2: Dimensions of Size 1 Are Repeated without Copying

````python
c = a + b
# c.shape == (3,5)
````

````python
print(c)
````

````python
c = a + b
# is logically equivalent to...
tmp_b = b.reshape(1,5)
tmp_b_repeat = tmp_b.repeat(3, 
                            axis=0)
c = a + tmp_b_repeat
# But broadcasting makes 
# no copies of 
# b's data
````

````python
print(c)
````

The trailing axes of either arrays must be 1 or both must have the same size for broadcasting to occur. Otherwise, a **"ValueError: shape mismatch: objects cannot be broadcast to a single shape"** exception is thrown.

## Broadcasting in Action

```
a = array([0,10,20,30])
b = array([0,1,2])
y = a[:,newaxis] + b
```

## Application: Distance from Centre

````python
import matplotlib.pyplot as plt
a = np.linspace(0,1,15) - 0.5
b = a[:, np.newaxis]
# b.shape == (15, 1)
dist2 = a**2 + b**2
# broadcasting sum
dist = np.sqrt(dist2)
plt.imshow(dist)
plt.colorbar()
````

## Broadcasting's Usefulness

Broadcasting can often be used
to replace needless data replication inside a NumPy array expression.
******
**np.meshgrid()** - use **newaxis** appropriately in broadcasting expressions.
******
**np.repeat()** - broadcasting makes repeating an array along a dimension of size 1 unnecessary

## Universal Function Methods

The mathematical, comparative, logical, and bitwise operators *op* that take two arguments (binary operators) have special methods that operate on arrays:
1. op.reduce(a,axis=0)
2. op.accumulate(a,axis=0)
3. op.outer(a,b)
4. op.reduceat(a,indices)

### op.reduce()

**op.reduce()** applies **op** to all the elements in a 1-D array **a** reducing it to a single value.

#### Add Example

````python
a = np.array(
        [1,2,3,4]
    )
np.add.reduce(a)
````

#### String List Example

````python
a = np.array(
        ['ab','cd','ef'],
        dtype='object'
    )
np.add.reduce(a)
````

#### Logical Op Examples

````python
a = np.array(
        [1,1,0,1]
    )
np.logical_and.reduce(a)
````

````python
np.logical_or.reduce(a)
````

For multidimensional arrays, **op.reduce(a,axis)** applies **op** to the elements of **a** along the specified **axis**.
******
The resulting array has dimensionality one less than **a**.
******
The default value for **axis** is 0.

#### Summing up Each Row

````python
a = np.arange(3) + np.arange(0,40,10).reshape(-1,1)
````

````python
np.add.reduce(a,1)
````

#### Sum Columns by Default

````python
np.add.reduce(a)
````

### op.accumulate()

**op.accumulate()** creates a new array containing the intermediate results of the **reduce** operation at each element in **a**.

#### Add Example

````python
a = np.array([1,2,3,4])
np.add.accumulate(a)
````

#### String List Example

````python
a = np.array(
        ['ab','cd','ef'],
        dtype='object'
    )
np.add.accumulate(a)
````

#### Logical Op Examples

````python
a = np.array([1,1,0])
np.logical_and.accumulate(a)
````

````python
np.logical_or.accumulate(a)
````

### op.reduceat()

**op.reduceat(a,indices)** applies **op** to ranges in the 1-D array **a** defined by the values in **indices**.
******
The resulting array has the same length as **indices**.

````python
a = np.array(
        [0,10,20,30,40,50]
    )
indices = np.array([1,4])
np.add.reduceat(a, indices)
````

NB: 
1. For multidimensional arrays,
**reduceat()** is always applied along the last axis (sum of rows for 2-D arrays).
2. This is different from the default for **reduce()** and **accumulate()**.

### op.outer()

**op.outer(a,b)** forms all possible combinations of elements between **a** and **b** using **op**.
******
The shape of the resulting array results from concatenating the shapes of **a** and **b** (ORDER MATTERS).

## Flattening Arrays

### Flatten (Safe)

**a.flatten()** converts multi-dimensional array into a 1-D array.
******
The new array is a *copy* of the original data.

````python
# Create a 2D array
a = np.array(
        [
            [0,1],
            [2,3]
        ]
    )
````

````python
# Flatten out elements to 1D
b = a.flatten()
b
````

````python
a
````

````python
# Changing b does not change a
b[0] = 10
b
````

````python
a
````

### Ravel (Efficient)

**a.ravel()** is the same as **a.flatten()**, but returns a *reference (or view)* of the array if possible (i.e. the memory is contiguous).
******
Otherwise, the new array copies the data.

````python
# Flatten out elements to 1-D
b = a.ravel()
b
````

````python
# Changing b does change a
b[0] = 10
b
````

````python
a
````
