# 1. Introduction

**Axis:** Refers to the X-axis and Y-axis.

**Axes | `axes.Axes` :** Refers to an entire plot. As far as Matplotlib is concerned, the histogram we just made is an **axes object** — NOT a plot or graph! Axes is not the plural form of axis.

**Figure | `figure.Figure` :** Refers to your entire canvas/workspace. A figure can contain a single axes, or it can contain many.

**Plot | `.plot()` :** In Matplotlib code, this word is always a **method/verb**. When people use “plot” as a noun, they are actually referring to an axes object.

**Subplot(s) | `.subplots()` :** In Matplotlib code, this word is a **method/verb**. When people use “subplots” as a noun, they are actually referring to an array of axes objects.

**fig/ax/axs:** These are variable names that are frequently used to store an instance of `figure.Figure`, `axes.Axes`, or an array of `axes.Axes`.

# 2. Levels of Subplots

There are 3 levels of subplots:
1. Axes Level
2. Subplot Level
3. Subplots (Plural)

## 2.1 Axes Level

Use `axes` and `figure.add_axes` : when you want PRECISE and unlimited control over the exact placement and drawing of your axes (*plots*), and don't mind defining the exact dimensions of each one.

## 2.2 Subplot Level

Use `figure.add_subplot` and `pyplot.subplot`: When you want to place your axes by hand, in grids you define yourself. Use with `plt.gridspec` for custom layouts to make it look better.

## 2.3 Subplots (Plural)

Use `figure.subplots` or `pyplot.subplots`: When you just want some rows of basic subplots, quickly, and don't care to get too fancy.

# 3. Imports & Data

```python
# Imports
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
import pandas.util.testing
# Make 14 rows of fake data.
a = pd.util.testing.makeDataFrame().reset_index().drop(columns='index')
b = pd.util.testing.makeDataFrame().reset_index().drop(columns='index').rename(columns={'A':'E', 'B':'F', 'C':'G', 'D':'H'}) 
c = pd.util.testing.makeDataFrame().reset_index().drop(columns='index').rename(columns={'A':'I', 'B':'J', 'C':'K', 'D':'L'}) 
d = pd.util.testing.makeDataFrame().reset_index().drop(columns='index').rename(columns={'A':'M', 'B':'N', 'C':'O', 'D':'P'})
df = pd.concat([a, b, c, d], axis=1).drop(columns=['O', 'P'])
df.head(7)
```

```python
plt.hist(df['A'])
```

*Add more axes to the figure and plot histograms on them.*

```python
fig = plt.figure(figsize=(20, 12))
ax0 = fig.add_subplot(3, 4, 1)
ax0.hist(df['B'])
ax1 = fig.add_subplot(3, 4, 2).hist(df['C']);
```

```python
fig = plt.figure(figsize=(20, 12))
# Generate 15 plots in a grid with 3 rows, 5columns
for i, col in enumerate(df.columns):
    ax = fig.add_subplot(3, 5, i+1) 
    ax.hist(df[col])
plt.show()
```

```python
# from IPython.display import display
fig, axs = plt.subplots(nrows=3, ncols=5, figsize=(20, 12))
axs[0, 0].hist(df['A'])
axs[1, 2].hist(df['B'])
axs[2, 1].hist(df['C'])
fig.show()
# # Display the figure
#display(fig)
```

```python
fig, axes = plt.subplots(3, 5, figsize=(20, 12))
for col, ax in zip(df.columns, axes.flat):
    print(col, type(df[col]), type(ax))
```

```python
fig, axes = plt.subplots(3, 5, figsize=(20, 12), sharey=True)
for col, ax in zip(df.columns, axes.flat):
    df[col].hist(ax=ax)
    ax.set_title(col)
    
# delete the leftover axes
for ax in axes.flat[df.columns.size:]:
     ax.set_axis_off()
```
